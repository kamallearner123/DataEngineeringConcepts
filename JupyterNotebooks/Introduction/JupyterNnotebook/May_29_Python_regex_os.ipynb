```python
data = '''
+919876543210 
David, India

Working in xyz company
having mail addresssssss david@xyx.com

Company is working on Cloud computing.....

friends1@gmail.com
friends3@gmail.com
friends2@gmail.com
'''

# REGULAR EXPRESSIONS
import re # importing re module

parsed_data = re.findall(r"\d{10}", data) # findall(REGEX, string)
print(parsed_data)

parsed_data = re.findall(r"\w+@\w+\.com", data) # findall(REGEX, string)
print(parsed_data)

message = "Hello, you are selected for appearing Hackethon"
for mail in parsed_data:
    print(f"Sending message {message} to {mail}")
```
